```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")
    )
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "A is the first character of English alphabet",
        "B is the second character of English alphabet"
    ]
)
```
